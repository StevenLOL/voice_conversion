# 分析合成のテスト

```python
WAV_DIR = "wav/clb"
target = "arctic_a0001"
```

```python
import os
import numpy as np
import subprocess
import matplotlib.pyplot as plt

def extract_pitch(wav_file, pitch_file):
    cmd = "bcut +s -s 22 %s | x2x +sf | pitch -a 1 -s 16 -p 80 > %s" % (wav_file, pitch_file)
    subprocess.call(cmd, shell=True)

def extract_mcep(wav_file, mcep_file):
    cmd = "bcut +s -s 22 %s | x2x +sf | frame -p 80 | window | mcep -m 25 -a 0.42 > %s" % (wav_file, mcep_file)
    print cmd
    subprocess.call(cmd, shell=True)

def draw_pitch(pitch_file, fs=16000):
    # バイナリファイルをダンプ
    cmd = "dmp +f %s > temp.txt" % pitch_file
    subprocess.call(cmd, shell=True)

    # ピッチを読み込む
    data = np.loadtxt("temp.txt")
    # Hzに変換
    for i in range(len(data)):
        if data[i, 1] != 0: data[i, 1] = fs / data[i, 1]            
    subprocess.call("rm -rf temp.txt")
    
    # 描画
    plt.plot(data[:,1])
    plt.xlabel("Frame")
    plt.ylabel("Hz")
    plt.show()
    
def draw_mcep(wav_file, mcep_file, frame):
    # スペクトルを求める
    cmd = "bcut +s -s 22 %s | x2x +sf | frame -p 80 | bcut +f -l 256 -s %d -e %d | window | spec | dmp +f > spec.txt" % (wav_file, frame, frame)
    print cmd
    subprocess.call(cmd, shell=True)
    
    # メルケプストラムをスペクトルに変換
    cmd = "bcut +f -n 25 -s %d -e %d < %s | mgc2sp -m 25 -a 0.42 | dmp +f > mcep.txt" % (frame, frame, mcep_file)
    subprocess.call(cmd, shell=True)

    # スペクトルとメルケプストラムを描画
    spec = np.loadtxt("spec.txt")
    plt.plot(spec[:, 1])
    mcep = np.loadtxt("mcep.txt")
    plt.plot(mcep[:, 1])
    
    plt.xlabel("frequency bin")
    plt.ylabel("log magnitude")
    plt.show()
    
    subprocess.call("rm -rf spec.txt")
    subprocess.call("rm -rf mcep.txt")

def synthesis(pitch_file, mcep_file, wav_file):
    cmd = "excite -p 80 %s | mlsadf -m 25 -a 0.42 -p 80 %s | clip -y -32000 32000 | x2x +fs > temp.raw" % (pitch_file, mcep_file)
    subprocess.call(cmd, shell=True)
    cmd = "sox -e signed-integer -c 1 -b 16 -r 16000 temp.raw %s" % (wav_file)
    subprocess.call(cmd, shell=True)
    cmd = "rm -rf temp.raw"
    subprocess.call(cmd, shell=True)
```

```python
wav_file = os.path.join(WAV_DIR, target + ".wav")
pitch_file = target + ".pitch"
mcep_file = target + ".mcep"
syn_file = target + ".syn.wav"
```

```python
# 分析再合成
extract_pitch(wav_file, pitch_file)
extract_mcep(wav_file, mcep_file)
draw_pitch(pitch_file)
draw_mcep(wav_file, mcep_file, 100)
synthesis(pitch_file, mcep_file, syn_file)
```

# GMMの学習

```python
import os
import numpy as np
from sklearn import mixture
from sklearn.utils import shuffle
from sklearn.externals import joblib
import matplotlib.pyplot as plt
```

```python
mcepDir1 = "./aligned_mcep_clb_slt/clb"
mcepDir2 = "./aligned_mcep_clb_slt/slt"
```

```python
# ランダムにnumTraining文を選択
numTraining = 50
mcepFiles = os.listdir(mcepDir1)
targetFiles = shuffle(mcepFiles)[:numTraining]
```

```python
# Joint Vectorsを作成

# ダミー行（vstackで追加するためのベース）
X = np.zeros((1, 26))
Y = np.zeros((1, 26))

# mcepファイルをロード
for mcepFile in targetFiles:
    mcep1 = np.loadtxt("%s/%s" % (mcepDir1, mcepFile))
    mcep2 = np.loadtxt("%s/%s" % (mcepDir2, mcepFile))
    X = np.vstack((X, mcep1))
    Y = np.vstack((Y, mcep2))

# ダミー行を削除
X = X[1:, :]
Y = Y[1:, :]
print X.shape, Y.shape

Z = np.hstack((X, Y))
print Z.shape

# バイナリ形式で保存しておく
np.save("Z.npy", Z)
```

```python
# 学習データのロード
Z = np.load("Z.npy")
print Z.shape

# 混合ガウスモデルの学習
g = mixture.GMM(n_components=32, covariance_type='full')
%time g.fit(Z)
print g

# モデルをファイルに保存
joblib.dump(g, 'clb_slt.gmm')
```

# 声質変換

```python
from scipy.stats import multivariate_normal

# 混合数
M = 32

d = 26

# 式(9)の計算
def P(m, x, g):
    M = g.weights_.shape[0]
    if m >= M:
        print "ERROR: out of range: %d >= %d" % (m, M)
        exit()
    
    # 分母を計算
    denom = 0
    for n in range(M):
        denom += g.weights_[n] * multivariate_normal.pdf(x, g.means_[n, 0:d], g.covars_[n, 0:d, 0:d])
    
    nume = g.weights_[m] * multivariate_normal.pdf(x, g.means_[m, 0:d], g.covars_[m, 0:d, 0:d])

    return nume / denom

# 式(11)の計算
def E(m, x):
    first = g.means_[m, d:]
    second = np.dot(g.covars_[m, d:, 0:d], np.linalg.inv(g.covars_[m, 0:d, 0:d]))
    return first + np.dot(second, x - g.means_[m, 0:d])

# 式(13)の計算
def convert(x):
    res = 0.0
    for m in range(M):
        res += P(m, x, g) * E(m, x)
    return res

print X[0]
print convert(X[0])
print Y[0]

target = 6
plt.plot(X[target], label="source")
plt.plot(convert(X[target]), label="convert")
plt.plot(Y[target], label="dest")
plt.legend()
```

```python
import subprocess

def extract_binary_mcep(wav_file, mcep_file):
    cmd = "bcut +s -s 22 %s | x2x +sf | frame -p 80 | window | mcep -m 25 -a 0.42 > %s" % (wav_file, mcep_file)
    subprocess.call(cmd, shell=True)

def draw_mcep_at_frame(wav_file, mcep_file, frame):
    # frame位置でのスペクトルを求める
    cmd = "bcut +s -s 22 %s | x2x +sf | frame -p 80 | bcut +f -l 256 -s %d -e %d | window | spec | dmp +f > spec.txt" % (wav_file, frame, frame)
    subprocess.call(cmd, shell=True)
    
    # frame位置でのメルケプストラムをスペクトルに変換
    cmd = "bcut +f -n 25 -s %d -e %d < %s | mgc2sp -m 25 -a 0.42 | dmp +f > mcep.txt" % (frame, frame, mcep_file)
    subprocess.call(cmd, shell=True)

    # スペクトルとメルケプストラムを描画
    spec = np.loadtxt("spec.txt")
    plt.plot(spec[:, 1])

    mcep = np.loadtxt("mcep.txt")
    plt.plot(mcep[:, 1], "r-", lw=2)
    
    plt.xlabel("frequency bin")
    plt.ylabel("log magnitude")
    plt.show()

frame = 50

wav_file = "wav/clb/arctic_a0005.wav"
mcep_file = "clb_a0005.mcep"
extract_binary_mcep(wav_file, mcep_file)
draw_mcep_at_frame(wav_file, mcep_file, frame)

wav_file = "wav/slt/arctic_a0005.wav"
mcep_file = "slt_a0005.mcep"
extract_binary_mcep(wav_file, mcep_file)
draw_mcep_at_frame(wav_file, mcep_file, frame)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def draw_mcep(mcep_file1, mcep_file2, loc, order):
    mcep1 = np.loadtxt(mcep_file1)
    mcep2 = np.loadtxt(mcep_file2)
    plt.figure(figsize=(8, 8))
    plt.subplot(loc)
    plt.plot(mcep1[:, order], label=mcep_file1)
    plt.plot(mcep2[:, order], label=mcep_file2)
    plt.xlabel("frame")
    plt.title("%dth-order mcep" % order)
    plt.tight_layout()
```
